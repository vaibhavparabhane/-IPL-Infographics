## Top 5 **Batsman**

```python
# Calculate total runs and innings played by each batter
batsman_stats = df.groupby('batter')['batsman_runs'].agg(['sum', 'count']).reset_index()
batsman_stats.columns = ["batter", "total_runs", "total_innings"]  # Renaming columns
```

```python
# Calculate total dismissals for each batter
player_dismissals = df.groupby('batter')['is_wicket'].sum().reset_index()
player_dismissals.columns = ["batter", "total_dismissals"]  # Renaming columns
```

```python
# Calculate the number of matches each batter has played
total_matches_played =df.groupby('batter')['match_id'].nunique().reset_index()
total_matches_played.columns = ["batter", "total_matches_played"]  # Renaming columns
```

```python
# Combine into single DataFrame
player_stats = (
    batsman_stats
    .merge(player_dismissals, on="batter", how="left")
    .merge(total_matches_played, on="batter", how="left")
)

# Display
player_stats
```

```python
#Calculate batting average
player_stats["average"] = player_stats["total_runs"] / player_stats["total_dismissals"]
```

```python
# Handle cases where total dismissals are 0.
player_stats["average"] = player_stats["average"].replace(np.inf, np.nan)
```

```python
# Sort players by batting average in descending order and pick the top 5
top_five_players = player_stats.sort_values(by="average", ascending=False).head(10)

# Display the top 10 players with the highest batting averages
top_five_players
```

## ***Top 5 Bowlers***

```python
# Calculate total wickets taken by each bowler
bowler_wickets = df.groupby('bowler')['is_wicket'].sum().reset_index()
bowler_wickets.columns = ["bowler", "total_wickets"]
```

```python
# Calculate total runs conceded by each bowler
bowler_runs = df.groupby('bowler')['total_runs'].sum().reset_index()
bowler_runs.columns = ["bowler", "total_runs_conceded"]
#Total runs conceded: The total number of runs given away by a team in all matches.
```

```python
# Merge the wickets and runs conceded data
bowler_stats = pd.merge(bowler_wickets, bowler_runs, on="bowler", how="left")
bowler_stats
```

```python
# Calculate bowling average: runs conceded per wicket
bowler_stats["bowling_average"] = bowler_stats["total_runs_conceded"] / bowler_stats["total_wickets"]
```

```python
# Handle cases where total wickets are 0 (to avoid infinite averages)
bowler_stats["bowling_average"] = bowler_stats["bowling_average"].replace(np.inf, np.nan)
```

```python
# Sort bowlers by bowling average in ascending order (lower is better) and pick the top 5
top_five_bowlers = bowler_stats.sort_values(by="bowling_average").head(5)
```

```python
# Display the top 5 bowlers with the best bowling averages
top_five_bowlers
```
